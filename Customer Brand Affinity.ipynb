```python
import pandas as pd
import numpy as np
import os

try:
    

    df = pd.read_excel('/Users/rajeshpillai/Downloads/Online_Retail.xlsx',skiprows=0,header=0)

    df = df.dropna()

except:
    print('File Not Found')
    
# Add Total by multiplying Quantity and Unit Price

df['Total'] = df.Quantity*df.UnitPrice

# Change date format to YYYYMM

df['DateYMD'] = df.InvoiceDate.dt.strftime('%Y%m%d')

df_customer = df.copy(deep=True)

df.head()
```

#### Define a dataframe by doing the following operation

Select custid, prod#, count(prod#)
from table
groupby custid, prod#
order by count(prod#)

And select the top 5 records

```python
df_customer = df.copy()

df_cust_prod = df_customer[['InvoiceNo','StockCode','CustomerID']]

df_cust_prod = df_cust_prod.groupby(by=['CustomerID','StockCode']).agg(
                                            {
                                            'StockCode':['count']
                                            })

df_cust_prod = df_cust_prod.rename(columns={"StockCode":"Code"})\
                           .droplevel(level=0,axis=1)\
                           .reset_index()\
                           .sort_values(by=['CustomerID','count'], ascending=False)

df_cust_prod.head()





```

```python
df_cust_prod_top3 = df_cust_prod.groupby('CustomerID').head(3)
```
