## Pareto Analysis

The objective of this analysis is to identify the Customers, Products and Stores(Locations) that contribute the sizable chunk of business.

This analysis will help us to understand the customers who need to be focused, Products which have growth potential, stores which have highest revenue generation etc



```python
import pandas as pd
import numpy as np
import os

try:
    

    df = pd.read_excel('/Users/rajeshpillai/Downloads/Online_Retail.xlsx',skiprows=0,header=0)

    df = df.dropna()

except:
    print('File Not Found')
    
# Add Total by multiplying Quantity and Unit Price

df['Total'] = df.Quantity*df.UnitPrice

# Change date format to YYYYMM

df['DateYMD'] = df.InvoiceDate.dt.strftime('%Y%m%d')

df_customer = df.copy(deep=True)

df.head()
```

### Customers

1. Find the number of customers

2. Total Sales per customers

3. Cumulative Sales per customer (in descending order)





```python
df_customer = df.copy(deep=True)

df_customer = df_customer[['CustomerID','Total']]

df_customer = df_customer.groupby(by='CustomerID').sum().sort_values(by='Total', ascending=False).reset_index()

# Remove all negative transactions

df_customer = df_customer[df_customer.Total >0]

# No of customers

no_of_customers = len(df_customer.CustomerID.unique())

df_customer.head()
```

df['cumulative_sum'] = df.total.cumsum()
df['cumulative_perc'] = 100*df.cumulative_sum/df.total.sum()



```python
# Cumulative Sum

df_customer['Cumulative_sum'] = df_customer.Total.cumsum()

df_customer['Cumulative_perc'] = 100 * df_customer.Cumulative_sum /df_customer.Total.sum()



df_customer.head()
```

```python
df_customer[df_customer.Cumulative_perc <80.0]
```
