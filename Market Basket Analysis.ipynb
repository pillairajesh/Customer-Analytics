```python
import pandas as pd
import numpy as np
import os

try:
    

    df = pd.read_excel('/Users/rajeshpillai/Downloads/Online_Retail.xlsx',skiprows=0,header=0)

    df = df.dropna()

except:
    print('File Not Found')
    
# Add Total by multiplying Quantity and Unit Price

df['Total'] = df.Quantity*df.UnitPrice

# Change date format to YYYYMM

df['DateYMD'] = df.InvoiceDate.dt.strftime('%Y%m%d')

df_customer = df.copy(deep=True)

df.head()
```

```python
# We want to see data when there is a sale of atleast 5 items together

df_mba = df[['InvoiceNo','StockCode']]

mba_count= df_mba.groupby('InvoiceNo')['StockCode'].count().sort_values(ascending=False)

mba_count.describe()

# We can remove the outlier scenarios and focus on the 3-12 range


df_mba_count = mba_count.to_frame()\
                        .reset_index()\
                        .rename(columns={"StockCode":"Count"})

#df_mba_count=df_mba_count[(df_mba_count.Count>3) & (df_mba_count.Count<25)]

df_mba_count=df_mba_count[df_mba_count.Count==12]

df_mba_count.head()

k = df_mba_count.InvoiceNo

df_mba = df_mba[df_mba.InvoiceNo.isin(k)]

df_mba['StockCode'] = df_mba['StockCode'].astype(str)

df_mba.head()
```

```python
df_mba = df_mba.groupby('InvoiceNo')['StockCode'].apply(tuple)

df_mba = df_mba.drop(columns=['InvoiceNo'])

df_mba.head()

print(len(df_mba))
```
